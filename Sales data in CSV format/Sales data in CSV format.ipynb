```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Sales data in CSV format
data = pd.read_csv('sample_sales_data.csv')

# Chart 1: Monthly Sales Bar Chart
data['Date'] = pd.to_datetime(data['Date'])
data['Month'] = data['Date'].dt.strftime('%Y-%m')
monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()

plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.bar(monthly_sales['Month'], monthly_sales['Sales'])
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Data')
plt.xticks(rotation=45)

# Chart 2: Daily Sales Line Chart
daily_sales = data.groupby('Date')['Sales'].sum().reset_index()

plt.subplot(132)
plt.plot(daily_sales['Date'], daily_sales['Sales'])
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Daily Sales Trends')
plt.xticks(rotation=45)

# Chart 3: Sales Distribution by Product Category Pie Chart
product_sales = data.groupby('Product')['Sales'].sum().reset_index()

plt.subplot(133)
plt.pie(product_sales['Sales'], labels=product_sales['Product'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Sales Distribution by Product Category')

plt.tight_layout()
plt.show()
```
