##  Exercise 1

Write a Python programming to display a bar chart of the popularity of programming Languages.

Sample data:

Programming languages: Java, Python, PHP, JavaScript, C#, C++

Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(languages, popularity)

plt.title('Popularity of Programming Languages')
plt.xlabel('Languages',color='blue')
plt.ylabel('Popularity',color='blue')

plt.show()
```

## Read face cream and facewash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


<img src="exercises_png/25.png">

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("datasets/company_sales_data.csv")
face_cream_sales = df ['facecream'].tolist()
face_wash_sales =  df ['facewash'].tolist()
months  = df ['month_number'].tolist()

#plt.bar(months, face_cream_sales, color='orange', label='Face Cream Sales Data')
#plt.bar(months, face_wash_sales, color='blue', label='Face Wash Sales Data')

plt.bar([a-0.25 for a in months], face_cream_sales, width= 0.25, label = 'Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in months], face_wash_sales, width= -0.25, label = 'Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(months)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

## Calculate total sale data for last year for each product and show it using a Pie chart

In Pie chart display Number of units sold per year for each product in percentage.

<img src="exercises_png/28.png">

```python
df = pd.read_csv("datasets/company_sales_data.csv")

products = ['Shampoo', 'Bathing soap', 'Moisturizer', 'Toothpaste','Facewash','Facecream']
salesData  = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(),df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]

plt.pie(salesData, labels=products, autopct='%1.1f%%')
plt.axis('equal')

plt.legend(loc='lower right')
plt.title('Sales Data')
plt.show()
```
