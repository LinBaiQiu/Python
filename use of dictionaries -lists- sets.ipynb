```python
# Creating a dictionary
student_info = {
    'name': 'John Doe',
    'age': 20,
    'major': 'Computer Science',
    'courses': ['Python Programming', 'Data Structures', 'Database Management'],
    'hobbies': {'reading', 'gaming', 'programming'}
}

# Accessing and printing dictionary elements
print("Student Information:")
print(f"Name: {student_info['name']}")
print(f"Age: {student_info['age']}")
print(f"Major: {student_info['major']}")
print("Courses:")

for course in student_info['courses']:
    print(f"- {course}")
print("Hobbies:", ', '.join(student_info['hobbies']))

# Modifying a dictionary
student_info['age'] = 21
student_info['courses'].append('Machine Learning')
student_info['hobbies'].add('playing chess')

# Printing modified dictionary
print("\nUpdated Student Information:")
print(f"Name: {student_info['name']}")
print(f"Age: {student_info['age']}")
print(f"Major: {student_info['major']}")
print("Courses:")

for course in student_info['courses']:
    print(f"- {course}")
print("Hobbies:", ', '.join(student_info['hobbies']))

# Converting the list of courses to a set
courses_set = set(student_info['courses'])

# Creating a list of student friends
friends = ['Alice', 'Bob', 'Charlie']

# Using a set to find common courses with a friend
friend_name = 'Alice'
friend_courses = ['Python Programming', 'Data Structures', 'Introduction to Psychology']
common_courses = courses_set.intersection(friend_courses)

print(f"\nCommon courses between {student_info['name']} and {friend_name}:")
for course in common_courses:
    print(f"- {course}")

# Using a list to find friends with the same major
print(f"\nFriends with the same major ({student_info['major']}):")
major_friends = [friend for friend in friends if friend != student_info['name']]
for friend in major_friends:
    print(f"- {friend}")
```

```python
# Create a dictionary to store student records
student_records = {}

# Add student records
student1 = {
    'name': 'Alice',
    'age': 20,
    'courses': ['Math', 'History', 'Physics'],
}
student2 = {
    'name': 'Bob',
    'age': 22,
    'courses': ['Computer Science', 'Statistics'],
}
student3 = {
    'name': 'Charlie',
    'age': 19,
    'courses': ['English', 'Psychology'],
}

student_records['alice_id'] = student1
student_records['bob_id'] = student2
student_records['charlie_id'] = student3

# Print student records
print("Student Records:")
for student_id, student_info in student_records.items():
    print(f"Student ID: {student_id}")
    print(f"Name: {student_info['name']}")
    print(f"Age: {student_info['age']}")
    print("Courses:")
    for course in student_info['courses']:
        print(f"- {course}")
    print()

# Create a list of courses available
courses_available = ['Math', 'Computer Science', 'Physics', 'Art', 'Statistics']

# Create a set of courses taken by students
courses_taken_by_students = set()
for student_info in student_records.values():
    courses_taken_by_students.update(student_info['courses'])

# Find courses that are not taken by any student
courses_not_taken = set(courses_available) - courses_taken_by_students

print("Courses not taken by any student:")
for course in courses_not_taken:
    print(f"- {course}")
```
